```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1D samples uniformly distributed in [-6, 6]
x = np.linspace(-6, 6, 200)
logistic_y = 1 / (1 + np.exp(-x))  # Logistic (sigmoid) mapping

# Create figure
fig, ax = plt.subplots(figsize=(6, 4))  # smaller figure

# Plot the logistic curve
ax.plot(x, logistic_y, color='#1E90FF', linewidth=2, label='Logistic Function σ(x)')

# Sample points (uniform in x)
num_points = 20
x_samples = np.linspace(-6, 6, num_points)
y_linear_samples = x_samples
y_logistic_samples = 1 / (1 + np.exp(-x_samples))

# Plot samples before and after transformation
ax.scatter(x_samples, np.zeros_like(x_samples), color='#2E8B57', s=35, label='Input Samples (x)')
ax.scatter(np.zeros_like(y_linear_samples) - 0.15, y_linear_samples / 12 + 0.5,
           color='#808080', s=35, label='Linear Mapping (scaled)')
ax.scatter(np.zeros_like(y_logistic_samples) + 0.15, y_logistic_samples,
           color='#D62728', s=35, label='Logistic Mapping')

# Add "L"-shaped guiding lines
for xi, ys in zip(x_samples, y_logistic_samples):
    ax.plot([xi, xi], [0, ys], color='#F4A6A6', linestyle=':', alpha=0.6)
    ax.plot([xi, 0.15], [ys, ys], color='#F4A6A6', linestyle=':', alpha=0.6)

# Axis limits
ax.set_xlim(-6.5, 6.5)
ax.set_ylim(-0.1, 1.1)

# Move labels inside the plot
ax.text(1, -0.08, "Input Space (x)", fontsize=10, color='dimgray', ha='right', va='bottom')
ax.text(-6.3, 0.7, "Output Space (y)", fontsize=10, color='dimgray', rotation=90, ha='left', va='top')

# Simplify grid and appearance
ax.grid(alpha=0.3)
ax.tick_params(axis='both', which='both', labelsize=8)
ax.set_xticks([-6, -3, 0, 3, 6])
ax.set_yticks([0, 0.5, 1])

# Legend — compact position inside plot
ax.legend(loc='upper left', bbox_to_anchor=(0.05, 0.9), frameon=False, fontsize=9)

# Tight layout and smaller margins
plt.subplots_adjust(left=0.08, right=0.97, top=0.95, bottom=0.12)
plt.tight_layout(pad=0.1)

plt.show()
```
